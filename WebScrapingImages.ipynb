```python
import requests
from bs4 import BeautifulSoup
import os
from tqdm import tqdm
from PIL import Image
from PIL import ImageFile
ImageFile.LOAD_TRUNCATED_IMAGES = True
```

```python
base_path=os.getcwd()+"/"
```

```python
for chap in tqdm(range(1,351)):
    URL = "https://ww5.mangakakalot.tv/chapter/manga-jn986670/chapter-"+str(chap) #chapter url
    page = requests.get(URL)
    soup=BeautifulSoup(page.text,'html.parser')
    images=soup.find_all('img')
    l=[] #list of image URLS
    i=1
    for image in range(len(images)):
        if(images[image]["alt"][:3]=="One"):
            l.append((images[image]['data-src']))   
    g=1
    k=[]
    for image in images:
        for i in range(len(l)):
            if str(l[i]) not in k:
                k.append(str(l[i]))
                name=image['alt']
                link=l[i]
                with open(name.replace(' ',"-").replace('/','')+".jpg",'wb') as f:
                    im=requests.get(link)
                    f.write(im.content)
                    f.close()
                    os.rename(base_path+"Manga-Online.jpg",str(chap)+"-"+str(g)+".jpg")
                    img = Image.open(str(chap)+"-"+str(g)+".jpg")
                    img.save("/content/drive/MyDrive/ALL programs/Face Detection Manga Characters/Data/"+str(chap)+"-"+str(g)+".jpg")
                    g=g+1
```
